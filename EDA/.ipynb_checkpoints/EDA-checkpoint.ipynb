There are two types of dataset: jpeg and Tensorflow Records.  
I personally prefer to Pytorch, I'm going to read jpeg files.  

```python
import torchvision.datasets as dset
from torchvision.datasets import VisionDataset
from torchvision import transforms
import torch
import matplotlib.pyplot as plt
from PIL import Image
import os
from pathlib import Path
```

```python
sample_image = Image.open('../dataset/gan-getting-started/monet_jpg/000c1e3bff.jpg')
```

```python
sample_image
```

```python
class monet_data(VisionDataset):
    def __init__(self, img_dir, transform=None):
        self.img_dir = Path(img_dir)
        self.images = []
        if transform == None:
            self.transform = transforms.ToTensor()
        else:
            self.transform = transform
        
    def setup(self):
        for i in os.listdir(self.img_dir):
            img = Image.open(self.img_dir/i)
            img = self.transform(img)
            self.images.append(img)
            
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        return self.images[idx]

dataset = monet_data('../dataset/gan-getting-started/monet_jpg/')
```

```python
dataset.setup()
```

```python
dataloader = torch.utils.data.DataLoader(dataset)
```

```python
batch = next(iter(dataloader))
```
